```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
from nltk.corpus import brown
import nltk
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline

nltk.download('brown')
```

```python
model = Word2Vec(sentences=brown.sents(), vector_size=3, window=5, min_count=1, workers=16)
model.save("word2vec.model")
```

```python
fig = plt.figure(figsize = (8,8))
# ax = fig.add_subplot(plt.axes())
ax = fig.add_subplot(plt.axes(projection='3d'))
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-1, 3)
```

```python
def plot(key):
    ax.scatter(* model.wv[key])
    ax.text(* model.wv[key], key)
```

```python
keys = ["Man", "Boy", "Guy", "Female", "male", "Woman", "Girl", "Gal",
        "Apple", "Banana", "Cherry",
        "person", "people",
        "Men", "Human", "humans",
        "We", "Us", "her", "she","he", "him", "they", "them"]

for k in keys:
    # plot(k)
    plot(k.lower())
fig
```
